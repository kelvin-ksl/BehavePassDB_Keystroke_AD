```python
import pandas as pd
from json_normalize import json_normalize
from datetime import datetime
from functools import reduce
from pytz import timezone
import json
import pytz
import os
```

```python
with open('TestSet_Task1_keystroke_enrolment.json') as f1:
      data1 = json.load(f1)
```

```python
df1 = pd.DataFrame(data1)
df_series1 = df1.squeeze()
```

```python
df_series1.index[[0]][0]
```

```python
vet1 = []
vet2 = []
vet3 = []
vet4 = []
vet5 = []
vet6 = []

vet1.clear()
vet2.clear()
vet3.clear()
vet4.clear()
vet5.clear()
vet6.clear()
```

```python
for i in df_series1.index[[0]]:
    print(i)
    index = df_series1[i]
    for linha in range(0, len(index['sensor_acc'])):
        pos1_sensor_acc = index['sensor_acc'][linha][0]
        pos2_sensor_acc = index['sensor_acc'][linha][1]
        pos3_sensor_acc = index['sensor_acc'][linha][2]
        pos4_sensor_acc = index['sensor_acc'][linha][3]
        
        #mostrando que pegou apenas um usuário
        user = df_series1.index[[0]][0]
        
        #criando o dicionario e passando para o vetor para ser criado o dataframe
        dicio_sensor_acc = {'sensor_acc': user, 'timestamp': pos1_sensor_acc, 'x_coordinate_sensor_acc': pos2_sensor_acc,'y_coordinate_sensor_acc': pos3_sensor_acc, 'z_coordinate_sensor_acc': pos4_sensor_acc}
        
        vet1.append(dicio_sensor_acc)
    
    for linha2 in range(0, len(index['sensor_grav'])):
        pos1_sensor_grav = index['sensor_grav'][linha2][0]
        pos2_sensor_grav = index['sensor_grav'][linha2][1]
        pos3_sensor_grav = index['sensor_grav'][linha2][2]
        pos4_sensor_grav = index['sensor_grav'][linha2][3]
        
        #mostrando que pegou apenas um usuário
        user = df_series1.index[[0]][0]
        
        dicio_sensor_grav = {'sensor_grav': user, 'timestamp': pos1_sensor_grav, 'x_coordinate_sensor_grav': pos2_sensor_grav,'y_coordinate_sensor_grav': pos3_sensor_grav, 'z_coordinate_sensor_grav': pos4_sensor_grav}
        
        vet2.append(dicio_sensor_grav)
    

    for linha3 in range(0, len(index['sensor_gyro'])):
        pos1_sensor_gyro = index['sensor_gyro'][linha3][0]
        pos2_sensor_gyro = index['sensor_gyro'][linha3][1]
        pos3_sensor_gyro = index['sensor_gyro'][linha3][2]
        pos4_sensor_gyro = index['sensor_gyro'][linha3][3]
         
        #mostrando que pegou apenas um usuário
        user = df_series1.index[[0]][0]
        
        dicio_sensor_gyro = {'sensor_gyro': user, 'timestamp': pos1_sensor_gyro, 'x_coordinate_sensor_gyro': pos2_sensor_gyro,'y_coordinate_sensor_gyro': pos3_sensor_gyro, 'z_coordinate_sensor_gyro': pos4_sensor_gyro}
        
        vet3.append(dicio_sensor_gyro)

    for linha4 in range(0, len(index['sensor_accl'])):
        pos1_sensor_accl = index['sensor_accl'][linha4][0]
        pos2_sensor_accl = index['sensor_accl'][linha4][1]
        pos3_sensor_accl = index['sensor_accl'][linha4][2]
        pos4_sensor_accl = index['sensor_accl'][linha4][3]
         
        #mostrando que pegou apenas um usuário
        user = df_series1.index[[0]][0]
        
        dicio_sensor_accl = {'sensor_accl': user, 'timestamp': pos1_sensor_accl, 'x_coordinate_sensor_accl': pos2_sensor_accl,'y_coordinate_sensor_accl': pos3_sensor_accl, 'z_coordinate_sensor_accl': pos4_sensor_accl}
        
        vet4.append(dicio_sensor_accl)     
    
    for linha5 in range(0, len(index['sensor_magn'])):
        pos1_sensor_magn = index['sensor_magn'][linha5][0]
        pos2_sensor_magn = index['sensor_magn'][linha5][1]
        pos3_sensor_magn = index['sensor_magn'][linha5][2]
        pos4_sensor_magn = index['sensor_magn'][linha5][3]
         
        #mostrando que pegou apenas um usuário
        user = df_series1.index[[0]][0]
        
        dicio_sensor_magn = {'sensor_magn': user,'timestamp': pos1_sensor_magn, 'x_coordinate_sensor_magn': pos2_sensor_magn,'y_coordinate_sensor_magn': pos3_sensor_magn, 'z_coordinate_sensor_magn': pos4_sensor_magn}
        
        vet5.append(dicio_sensor_magn)    

    for linha6 in range(0, len(index['touch'])):
        #parte do sensor de touch
        pos1_sensor_touch = index['touch'][linha6][0]
        pos2_sensor_touch = index['touch'][linha6][1]
        
        #mostrando que pegou apenas um usuário
        user = df_series1.index[[0]][0]
        
        dicio_sensor_touch = {'touch': user, 'timestamp': pos1_sensor_touch, 'ascii_code_touch': pos2_sensor_touch}
    
        vet6.append(dicio_sensor_touch)
```

```python
df_sensor_acc_one_user = pd.DataFrame(vet1)
df_sensor_acc_one_user.drop(columns='sensor_acc',inplace=True)
df_sensor_acc_one_user
```

```python
df_sensor_grav_one_user = pd.DataFrame(vet2)
df_sensor_grav_one_user.drop(columns='sensor_grav',inplace=True)
df_sensor_grav_one_user.head()
```

```python
df_sensor_gyro_one_user = pd.DataFrame(vet3)
df_sensor_gyro_one_user.drop(columns='sensor_gyro',inplace=True)
df_sensor_gyro_one_user.head()
```

```python
df_sensor_accl_one_user = pd.DataFrame(vet4)
df_sensor_accl_one_user.drop(columns='sensor_accl',inplace=True)
df_sensor_accl_one_user.head()
```

```python
df_sensor_magn_one_user = pd.DataFrame(vet5)
df_sensor_magn_one_user.drop(columns='sensor_magn',inplace=True)
df_sensor_magn_one_user.head()
```

```python
df_sensor_touch_one_user = pd.DataFrame(vet6)
df_sensor_touch_one_user.drop(columns='touch',inplace=True)
df_sensor_touch_one_user.head()
```

### ***Realizando o agrupamento e resample dos dados***

```python
# Convert timestamp to datetime object and set as index
df_sensor_acc_one_user['timestamp'] = pd.to_datetime(df_sensor_acc_one_user['timestamp'])#, unit='us') 
df_sensor_grav_one_user['timestamp'] = pd.to_datetime(df_sensor_grav_one_user['timestamp'])#, unit='us')
df_sensor_gyro_one_user['timestamp'] = pd.to_datetime(df_sensor_gyro_one_user['timestamp'])#, unit='us')
df_sensor_accl_one_user['timestamp'] = pd.to_datetime(df_sensor_accl_one_user['timestamp'])#, unit='us')
df_sensor_magn_one_user['timestamp'] = pd.to_datetime(df_sensor_magn_one_user['timestamp'])#, unit='us')
df_sensor_touch_one_user['timestamp'] = pd.to_datetime(df_sensor_touch_one_user['timestamp'])#, unit='us')


# Set index of columns
df_sensor_acc_one_user.set_index('timestamp', inplace=True)
df_sensor_grav_one_user.set_index('timestamp', inplace=True)
df_sensor_gyro_one_user.set_index('timestamp', inplace=True)
df_sensor_accl_one_user.set_index('timestamp', inplace=True)
df_sensor_magn_one_user.set_index('timestamp', inplace=True)
df_sensor_touch_one_user.set_index('timestamp', inplace=True)

# Exemple resample (2S), miliseconds (100L), Micro (100000us) and Nano (100000000N)
# Resample to a frequency of 100 milliseconds and calculate rolling mean with window of 10 samples

df_sensor_acc_one_user = df_sensor_acc_one_user.resample('100L').mean().rolling(window=10).mean()
df_sensor_grav_one_user = df_sensor_grav_one_user.resample('100L').mean().rolling(window=10).mean()
df_sensor_gyro_one_user = df_sensor_gyro_one_user.resample('100L').mean().rolling(window=10).mean()
df_sensor_accl_one_user = df_sensor_accl_one_user.resample('100L').mean().rolling(window=10).mean()
df_sensor_magn_one_user = df_sensor_magn_one_user.resample('100L').mean().rolling(window=10).mean()
df_sensor_touch_one_user = df_sensor_touch_one_user.resample('100L').mean().rolling(window=10).mean()


data_frames = [df_sensor_acc_one_user, df_sensor_grav_one_user, df_sensor_gyro_one_user, df_sensor_accl_one_user,
               df_sensor_magn_one_user, df_sensor_touch_one_user]

# Merge DataFrames
df_merged = reduce(lambda left,right: pd.merge(left,right , left_index=True, right_index=True), data_frames)

# Reset index
df_merged.reset_index(inplace=True)
```
